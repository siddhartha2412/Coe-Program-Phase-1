Objective:
Simulate basic ATM operations with validation and transaction rules.
1. User must enter a valid 4-digit PIN to access ATM features.
 Features & Rules:
PIN Validation
I. User must enter the correct 4-digit PIN to proceed.
I. Deposit
1. Amount must be in multiples of ₹100.
2. Maximum per transaction: ₹20,000.
II. Withdraw
1. Amount must be in multiples of ₹100.
2. Must not exceed account balance.
3. Maximum per transaction: ₹20,000.

III. Balance Enquiry
Display current balance to the user.

Assignment Note:
Implement each operation as a function, handle invalid inputs, and display clear messages. Initial balance can be assumed (e.g., ₹10,000).

```python
obj=Bank()
obj.validate()
class Bank:
    pin = "5100"
    acbal = 6000
    wcount = 0
    
    
    
    def validate(self):
        attempts =0
        while attempts < 3 :
            epin = input("Enter your pin")
            if epin == self.pin:
                print("Pin verified succesfully...")
                self.viewOptions()
            else :
                attempts +=1
                if attempts == 3:
                    break
                print("Invalid pin!!!")
                print("You have ",3-attempts,"attempts left !!!")
        print("you have exceeded your attempts for the day.\nplease wait for 24 hrs. or  visit to our bank nearby")
            
            
            
    def viewOptions(self):
        print("Enter the options")
        print("1.DEPOSIT")
        print("2.WITHDRAW")
        print("3.BALANCE ENQUIRY")
        print("4.EXIT")
        option = input()
        if option == "1" :
            self.deposit()
        elif option == "2":
            if self.wcount<3:
                self.withdraw()
            else:
                print("Your withdrawal limit for the day is over. Please try again after 24 hours.")
        elif option == "3":
            self.balance_enquiry()
        else :
            self.exit()
            
            
            
    def deposit(self):
        self.amount = int(input("Enter amount you want to deposit"))
        if self.amount%100 == 0:
            if self.amount<=20000:
                self.acbal += self.amount
                print("Money deposited successfully")
                print("Current balance is ",self.acbal)
            else :
                print("The amount should be less than 20000")
        else :
            print("Enter amount in multiples of ₹100.")
            
            
            
    def withdraw(self):
        amount = int(input("Enter amount you want to withdraw"))
        if amount%100 == 0:
            if amount<=self.acbal :
                self.acbal -= amount
                self.wcount += 1
                self.calculate_notes(amount)
                print("")
                print("Current balance is ",self.acbal)
                self.viewOptions()
            else :
                print("Insufficient balance")
        else :
            print("amount should be in multiples of 100")
    def calculate_notes(self,amount):
        notes_500 = amount//500
        amount -= notes_500*500 
        notes_200 = amount//200
        amount -= notes_200*200
        notes_100 = amount//100
        amount -= notes_100*100
        print("500 :",notes_500)
        print("200 :",notes_200)
        print("100 :",notes_100)
    def balance_enquiry(self):
        print("current balance is ",self.acbal)
        self.viewOptions()
        
    def exit(self):
        print("Are you sure want to exit?")
        option = input("1.Yes\n2.No")
        if option=="1":
            print("Thank you , please visit again!!!")
        else :
            self.viewOptions()
```
